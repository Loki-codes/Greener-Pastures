```python
#Importing GEO ID's
covid_info = "data/COVID - City - Daily.csv"
covid_info_df = pd.read_csv(covid_info)
covid_info_df
```

```python
#replacing all . with NaN
covid_info_df = covid_info_df.replace({'.': "NaN"})
covid_info_df
```

```python
#dropping un-needed columns
covid_info_df = covid_info_df.drop(['case_count',
                        'death_count',
                       'case_rate',
                       'death_rate',
                       'new_death_rate',
                       'new_case_count',
                       'new_death_count'], axis = 1)
covid_info_df
```

```python
#finding covid new case rate data for the 22nd of each month
day_22_covid = (covid_info_df['day'] == 22)
day_22_covid = covid_info_df[day_22_covid]
day_22_covid
```

```python
#***DO NOT RE RUN. RESTART AND CLEAR KERNEL TO RE RUN. just rerunning will keep moving the columns
#moving cityid to the left to math other dataframe
shiftcol = day_22_covid.columns.tolist()
shiftcol = shiftcol[3:] + shiftcol[:3]
day_22_covid = day_22_covid[shiftcol]
day_22_covid
```

```python
merge_cv_mrchnt = month_merge_df_reset.merge(day_22_covid, on=['cityid', 'year', 'month', 'day'])
merge_cv_mrchnt.head(45)
```
