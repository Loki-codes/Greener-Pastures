```python
import pandas as pd
from pprint import pprint
#importing the cost of living csv files
df1rent = pd.read_csv('./data/Rent_1br.csv')
df3rent = pd.read_csv('./data/Rent_3br.csv')
dfu = pd.read_csv('./data/Utilities.csv')

```

```python
#converting ther csv into data frames
df1bd = pd.DataFrame(df1rent, index = None) 
df3bd = pd.DataFrame(df3rent, index = None) 
dfUt = pd.DataFrame(dfu, index = None) 
```

```python
#merging the data frames into one data frame
adf = df1bd.merge(df3bd, how='outer', on='City',suffixes=('_1bd', '_3bd'))
df = adf.merge(dfUt, how='outer', on='City')
```

```python
# dropping unneeded columns
ndf = df[['City', 'Apartment_1bd', 'Apartment_3bd', 'Basic']]
#to keep New York City in the data set it was esiest to just rename it here
ndf.iloc[1,0] = 'New York City'
ndf.head()
```

```python
#Rename columns
ndf = ndf.rename(columns={"Apartment_1bd": "Cost of 1 Bedroom Appartment", "Apartment_3bd": 'Cost of 3 Bedroom Appartment',\
                  'Basic': 'Utilities Per Month'})
```

```python
#now to only keep data for the Cities I care about. I made another list with all of the cities\
#in the df in case I want to do an outer join of the lists to create a list of the cities that will be removed. 
cityList = ['Los Angeles', 'New York City', 'Chicago', 'Houston', 'Phoenix', 'San Diego',
 'Dallas', 'Las Vegas', 'Seattle', 'Fort Worth', 'San Antonio', 'San Jose',
 'Detroit', 'Philadelphia', 'Columbus', 'Austin', 'Charlotte', 'Indianapolis',
 'Jacksonville', 'Memphis', 'San Francisco', 'El Paso', 'Baltimore', 'Portland',
 'Boston', 'Oklahoma City', 'Louisville', 'Denver', 'Washington', 'Nashville',
 'Milwaukee', 'Albuquerque', 'Tucson', 'Fresno', 'Sacramento', 'Atlanta',
 'Kansas City', 'Miami', 'Raleigh', 'Omaha', 'Oakland', 'Minneapolis', 'Tampa',
 'New Orleans', 'Wichita', 'Cleveland', 'Honolulu', 'Boise',
 'Salt Lake City', 'Colorado Springs', 'Tulsa']
totalCities = ndf['City'].unique()
fdf =pd.DataFrame(columns=['City'])
for i in cityList: 
    for index, row in ndf.iterrows():  
        if (i) in row['City']:
            fdf = fdf.append(row)
             
            

```

```python
fdf.reset_index(drop = "True")
```

```python
# I had to make a n ew list, alphabatize them and zip them together before pretty printing them so I could see where \
#things were no meshing together perfectly. 
sdf = fdf.sort_values(by='City', ascending=True)
newCityL = sdf['City'].unique()
cityList.sort()
newCityL.sort()
nL = zip(cityList,newCityL)
pnL = list(nL)
#pprint(pnL)


            
```

```python
#Now that I know that the dataframe and the list of cities we are using as keys are the
#same I can make a new column and the final dataFrame.
sdf['City Name'] =  cityList
costoOfLivingDF = sdf[['City Name', 'Cost of 1 Bedroom Appartment', 'Cost of 3 Bedroom Appartment', 'Utilities Per Month']]
```
