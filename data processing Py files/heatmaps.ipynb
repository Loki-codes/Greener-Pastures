```python
#Defining Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import gmaps
import os

# Import API key
from key_for_heatmap import g_key
```

```python
#Importing GEO ID's
summ_geo_info = "../data/GeoIDs - City.csv"
summ_geo_info_df = pd.read_csv(summ_geo_info)
summ_geo_info_df.head()
```

```python
#Dropping un-need columns
geo_dropped = summ_geo_info_df.drop(columns=['statename', 'statefips'])
geo_dropped.set_index("cityid")
```

```python
#Importing Womply
summ_womply_info = "../data/Womply - City - Daily.csv"
summ_womply_info_df = pd.read_csv(summ_womply_info)
summ_womply_info_df.head()
```

```python
len(summ_womply_info_df)
```

```python
#Dropping un-needed columns
womply_dropped = summ_womply_info_df.drop(columns=['merchants_ss40', 'merchants_ss60',
                                 'merchants_ss65', 'merchants_ss70',
                                 'revenue_all','revenue_ss40',
                                  'revenue_ss60','revenue_ss65',
                                 'revenue_ss70'])
womply_dropped
```

```python
#Merging geo and womply df's
merge_df = pd.merge(geo_dropped, womply_dropped, on="cityid")
merge_df
```

```python
#finding the info only on the 22nd of the month
monthly_ = merge_df['day'] == 22
monthly_merge_df = merge_df[monthly_]
monthly_merge_df
```

```python
#merging by cityid
month_merge_df = monthly_merge_df.set_index("cityid")
month_merge_df
```

```python
#adding column to find difference
month_merge_df['Merchants Difference'] = ""
month_merge_df.head(50)
```

```python
month_merge_df_reset = month_merge_df.reset_index(drop=False)
month_merge_df_reset
```

```python
for city in month_merge_df_reset["cityname"].unique():
    month_merge_df_reset["Merchants Difference"][month_merge_df_reset["cityname"]==city] = month_merge_df_reset["merchants_all"][month_merge_df_reset["cityname"]==city].diff().tolist()
month_merge_df_reset.head(40)
```

```python
#find the month to month difference
#month_merge_df_reset= month_merge_df_reset.drop(month_merge_df_reset["Merchants Difference"]=="NaN")
#month_merge_df_reset["Merchants Difference"].fillna(0)
month_merge_df_reset["Difference plus 1"] = ((month_merge_df_reset["Merchants Difference"] + 1)*100)
month_merge_df_reset.head(40)
```

```python
diff_max= month_merge_df_reset["Difference plus 1"].max()
```

```python
gmaps.configure(api_key=g_key)
```

```python
heat_1_df = month_merge_df_reset[(month_merge_df_reset["month"]==2)&(month_merge_df_reset["year"]==2020)]
heat_1_df
```

```python
# Possible to do a zoom label, satellite image, get negative and posititve value in a single map.
#head_1_df = month_merge_df_reset[month_merge_df_reset["month"]==1]
location = heat_1_df[["lat","lon"]].astype(float)
merchant = heat_1_df["Difference plus 1"].astype(float)
```

```python
month_2= heat_1_df["Difference plus 1"].max()
```

```python

fig = gmaps.figure()


month_2_map = gmaps.heatmap_layer(location,weights=merchant,
                                dissipating=False,max_intensity=diff_max,
                                point_radius=1)



#heat_layer.gradient= ["blue","green","red"]
fig.add_layer(month_2_map)
fig
```
