```python
#Import section
import pandas as pd
# I pulled in the data concerning job postings on the app "Burnong Glass"
odf = pd.read_csv ('./data/Burning Glass - City - Weekly.csv')

#trimed down the columns to what I would need. 
tdf = odf[['year','month','day_endofweek','cityid','bg_posts']]
tdf['year_month']= ''

for index, row in tdf.iterrows():
    tdf.loc[index, "year_month"] = str(row["year"]) + ':' + str(row["month"])
    
# We only wanted the most recent month of data
df = tdf.drop_duplicates(subset=['year_month', 'cityid'], keep = "last")
fdf = df.loc[df['year_month']== '2020:2']
fdf.reset_index()
fdf.head(53)
```

```python
#Lastly I sent the file to our data repository.
fdf.to_csv(r'./data/Job Postings.csv', index = False)
```
