```python
#Defining Dependencies
import pandas as pd
```

```python
#Importing GEO ID's
summ_geo_info = "data/GeoIDs - City.csv"
summ_geo_info_df = pd.read_csv(summ_geo_info)
summ_geo_info_df.head()
```

```python
#Dropping un-need columns
geo_dropped = summ_geo_info_df.drop(columns=['statename', 'statefips'])
geo_dropped
```

```python
#Importing Womply
summ_womply_info = "data/Womply - City - Daily.csv"
summ_womply_info_df = pd.read_csv(summ_womply_info)
summ_womply_info_df.head()
```

```python
#Dropping un-needed columns
womply_dropped = summ_womply_info_df.drop(columns=['merchants_ss40', 'merchants_ss60',
                                 'merchants_ss65', 'merchants_ss70',
                                 'revenue_all','revenue_ss40',
                                  'revenue_ss60','revenue_ss65',
                                 'revenue_ss70'])
womply_dropped
```

```python
#Merging geo and womply df's
merge_df = pd.merge(geo_dropped, womply_dropped, on="cityid")
merge_df
```

```python
#finding the info only on the 22nd of the month
monthly_ = merge_df['day'] == 22
monthly_merge_df = merge_df[monthly_]
monthly_merge_df
```

```python
#merging by cityid
month_merge_df = monthly_merge_df.set_index("cityid")
month_merge_df
```

```python
#adding column to find difference
month_merge_df['Merchants Difference'] = ""
month_merge_df.head(50)
```
